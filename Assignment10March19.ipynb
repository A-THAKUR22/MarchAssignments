1. Sample problem of predicting home price in monroe, new jersey (USA)
Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Read the data from csv
data=pd.read_csv('homeprices (1).csv')
df=pd.DataFrame(data)
df
```

```python
#Replacing the NaN values with the median of the column
median_bedrooms = df['bedrooms'].median()
df['bedrooms'].fillna(median_bedrooms, inplace=True)
df
```

```python
#Split the data into features (X) and target variable (y)
X = df[['area', 'bedrooms', 'age']]
y = df['price']

#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)

#Predict on the basis of new values
new_values=[[3000,3,40]]
predictions=model.predict(new_values)
print(f'Predicted price for the house is : {predictions[0]}')
```

Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old.

```python
second_values_to_predict=[[2500,4,5]]
new_prediction=model.predict(second_values_to_predict)
print(f'Predicted price for the house is : {new_prediction[0]}')
```

2. There is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
#Read the data from csv file
n_data=pd.read_csv('hiring.csv')
df1=pd.DataFrame(n_data)
df1
```

```python
#Replace the NaN values with the median and mode in the particular columns
median_test=df1['test_score(out of 10)'].median()
mode_experience=df1['experience'].mode([0])
df1['test_score(out of 10)'].fillna(median_test,inplace=True)
df1['experience'].fillna(mode_experience,inplace=True)

#Using the label encoding to use appropriate values for the categorical data in 'experience' column
encoder=LabelEncoder()
df1['experience(encoded)']=encoder.fit_transform(df1['experience'])
df1.drop(columns=['experience'], inplace=True)
df1
```

2 yr experience, 9 test score, 6 interview score

```python
#Split the data into features (X) and target variable (y)
X=df1[['experience(encoded)','test_score(out of 10)','interview_score(out of 10)']]
y=df1['salary($)']

#Split the data into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.31,random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#Performing the predictions based on new entries
new_entry={
    'experience(encoded)':[encoder.transform(['two'])[0]],
    'test_score(out of 10)':[9],
    'interview_score(out of 10)':[6]
}
new_data=pd.DataFrame(new_entry)
predicted_salary=model.predict(new_data)
print(f'Predicted Salary in dollars is:{predicted_salary[0]}')
```

12 yr experience, 10 test score, 10 interview score

```python
second_entry={
    'experience(encoded)':[encoder.fit_transform(['twelve'])[0]],
    'test_score(out of 10)':[10],
    'interview_score(out of 10)':[10]
}

second_data=pd.DataFrame(second_entry)
second_predicted_salary=model.predict(second_data)
print(f'Predicted Salary in dollars: {second_predicted_salary[0]}')
```
